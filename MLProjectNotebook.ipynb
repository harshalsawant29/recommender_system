```python
import json
import csv

def jsonl_to_csv(jsonl_file, csv_file):
    with open(jsonl_file, 'r') as f:
        with open(csv_file, 'w', newline='', encoding='utf-8') as csvf:
            writer = csv.writer(csvf)
            # Write header
            writer.writerow(['rating', 'title', 'text', 'asin', 'parent_asin', 'user_id', 'timestamp', 'helpful_vote', 'verified_purchase'])
            # Process each line in the JSON Lines file
            for line in f:
                data = json.loads(line)
                writer.writerow([
                    data.get('rating', ''),
                    data.get('title', ''),
                    data.get('text', ''),
                    data.get('asin', ''),
                    data.get('parent_asin', ''),
                    data.get('user_id', ''),
                    data.get('timestamp', ''),
                    data.get('helpful_vote', ''),
                    data.get('verified_purchase', '')
                ])

# Example usage:
jsonl_file = 'Grocery_and_Gourmet_Food.jsonl'
csv_file = 'output.csv'
jsonl_to_csv(jsonl_file, csv_file)
```
